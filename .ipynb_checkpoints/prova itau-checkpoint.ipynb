```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data/regressao_Q1.csv")
```

```python
data.head()
```

```python
from sklearn.linear_model import ElasticNet
```

```python
regr = ElasticNet(alpha=1 , l1_ratio=0.01)
```

```python
from sklearn import metrics
```

```python
X = data.drop(columns=data.columns[-1],  axis=1,  inplace=False)
```

```python
X.head()
```

```python
y = data.iloc[:,-1:]
y.head()
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold, KFold
cv = cross_validate(regr, X, y, cv=KFold(5).get_n_splits(), scoring="neg_mean_squared_error", return_train_score=True)
```

```python
-np.mean(cv['train_score']),-np.mean(cv['test_score'])
```
